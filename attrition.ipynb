## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[['Attrition', 'Department']]
y_df.head()

```

```python
# Create a list of at least 10 column names to use as X data

selected_columns = ['Age', 'DistanceFromHome', 'HourlyRate', 'JobRole','JobLevel', 'NumCompaniesWorked', 'PercentSalaryHike', 'TotalWorkingYears', 'YearsAtCompany', 'YearsSinceLastPromotion']



# Create X_df using your selected columns
X_df = attrition_df[selected_columns]


# Show the data types for X_df
X_df.dtypes



```

```python
#chnage the data type of the JobRole column to int64
X_df['JobRole'] = X_df['JobRole'].astype('category').cat.codes
```

```python
# Split the data into training and testing sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, test_size=0.2, random_state=42)


```

```python
# Convert your X data to numeric data types however you see fit Add new code cells as necessary

X_df = X_df.astype('float64')
X_df.dtypes






```

```python
# Create a StandardScaler
scaler = StandardScaler()

# Fit the StandardScaler to the training data
scaler.fit(X_train)

# Scale the training and testing data
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)





```

```python
# Create a OneHotEncoder for the Department column
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse=False)

department = attrition_df[['Department']]
department_encoded = encoder.fit_transform(department)

encoder = OneHotEncoder(sparse=False)
y_train_encoded = encoder.fit_transform(y_train)
y_test_encoded = encoder.transform(y_test)


# Fit the encoder to the training data
enc = OneHotEncoder(sparse=False)

# Create two new variables by applying the encoder to the training and testing data

enc.fit(y_df)
y_train = enc.transform(y_df)
y_test = enc.transform(y_df)

y_train_encoded[:5]
y_test_encoded[:5]


```

```python
# Create a OneHotEncoder for the Attrition column
from sklearn.preprocessing import OneHotEncoder

attrition = attrition_df[['Attrition']]
attrition_encoded = encoder.fit_transform(attrition)

encoder = OneHotEncoder(sparse=False)
y_train_encoded = encoder.fit_transform(y_train)
y_test_encoded = encoder.transform(y_test)


# Fit the encoder to the training data
enc = OneHotEncoder(sparse=False)


# Create two new variables by applying the encoder
# to the training and testing data

enc.fit(y_df)
y_train = enc.transform(y_df)
y_test = enc.transform(y_df)

y_train_encoded[:5]
y_test_encoded[:5]

```

```python
# Define input layers
department_input = layers.Input(shape=(number_input_features,))
attrition_input = layers.Input(shape=(number_input_features,))

# Shared layers
shared_layer1 = layers.Dense(units=64, activation='relu')(department_input)  # Connect to department_input
shared_layer2 = layers.Dense(units=32, activation='relu')(shared_layer1)

# Department branch
department_hidden = layers.Dense(units=64, activation='relu')(shared_layer2)  # Connect to shared_layer2
department_output = layers.Dense(units=2, activation='relu')(department_hidden)

# Attrition branch
attrition_hidden = layers.Dense(units=64, activation='relu')(shared_layer2)  # Connect to shared_layer2
attrition_output = layers.Dense(units=2, activation='relu')(attrition_hidden)

# Create the model
model = Model(inputs=[department_input, attrition_input], outputs=[department_output, attrition_output])
```

```python
# Create the model
model = Model(inputs=[department_input, attrition_input], outputs=[department_output, attrition_output])

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Summarize the model
model.summary()


```

```python
# Train the model
model.fit([X_train_scaled, X_train_scaled], [y_train_encoded, y_train_encoded], epochs=10, batch_size=32)
```

```python
# Evaluate the model with the testing data
model_loss, model_accuracy = model.evaluate([X_test_scaled, X_test_scaled], [y_test_encoded, y_test_encoded])
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
